Multilabel spotting: spotting samples with more than one ground truth. Need to put them in the training set and not in the testing.

```python
#multilabel spotting. ids stored in multilabel
seen = {}
dupes = []
label = list()

for i in data_index.keys():
    for x in data_index[i]:
        if x not in seen:
            seen[x] = 1
        else:
            if seen[x] == 1:
                dupes.append(x)
            seen[x] += 1

multilabel = list()
for key, value in seen.items():
    if value > 1:
        multilabel.append(key)
```

```python
multilabel_dict = dict()
counter_multilabel = 0
for i in data_index.keys():
    multilabel_dict[i] = list()
    for x in data_index[i]:
        if x in multilabel:
            multilabel_dict[i].append(x)
            counter_multilabel = counter_multilabel + 1
```

We can see that the samples that are multilabelled are mostly cupboard/drawer/door opening and closing. It makes sense.

```python
pre_csv = dict()
for i in true_dataset.keys():
    pre_csv[i] = list()
    for j in true_dataset[i]:
        #don't consider the multilabel ones
        if j['id'] not in multilabel:
            j['filename'] = str(j['id']) + '.wav'
            pre_csv[i].append(j['filename'])
```

```python
df = pd.DataFrame(columns=['fname', 'label', 'manually_verified'])
```

```python
counter = 0
for i in pre_csv.keys():
    for j in pre_csv[i]:
        df.loc[counter] = [j, i, 1]
        counter = counter + 1
```

```python
# Prepare data for fitting classifier (as sklearn classifiers require)
y = list()
X = list()
for class_name, sounds in pre_csv.items():
    for count, sound in enumerate(sounds):
        # Use index of class name in class_names as numerical value (classifier internally represents
        # class label as number)
        y.append(class_names.index(class_name))
        #feature_vector = build_feature_vector(sound)
        X.append(sound)
```

```python
X_train_HQ, X_test, y_train_HQ, y_test = train_test_split(X, y, test_size=0.4)
```

Appending multilabel samples in the training set -> so we have only a multiclass problem and not multilabel

```python
for key in multilabel_dict.keys():
    for i in range(len(multilabel_dict[key])):
        X_train_HQ.append(str(multilabel_dict[key][i]) + '.wav')
        y_train_HQ.append(class_names.index(key))
```
