```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import utils

%matplotlib inline
```

```python
DATASET_NAME = 'baseline'
true_dataset = utils.load_from_json('%s.json' % DATASET_NAME)
print('Loaded dataset with %i classes:' % len(true_dataset))
for klass, sounds in true_dataset.items():
    print('\t%s: %i sounds' % (klass, len(sounds)))
class_names = list(true_dataset.keys())
```

```python
durations = [[]]
silence_20 = [[]]
zcr = [[]]
logattacktime = [[]]
inharmonicity = [[]]
```

```python
for key in true_dataset.keys():
    durations_cat = list()
    silence_20_cat = list()
    zcr_cat = list()
    logattacktime_cat = list()
    inharmonicity_cat = list()
    for i in range(len(true_dataset[key])):
        try:
            durations_cat.append(true_dataset[key][i]['analysis']['sfx']['duration'])
            silence_20_cat.append(true_dataset[key][i]['analysis']['lowlevel']['silence_rate_20dB']['mean'])
            #silence_30_dict[key] = sum_silence_30 / counter
            #silence_60_dict[key] = sum_silence_60 / counter
            zcr_cat.append(true_dataset[key][i]['analysis']['lowlevel']['zerocrossingrate']['mean'])
            logattacktime_cat.append(true_dataset[key][i]['analysis']['sfx']['logattacktime']['mean'])
            inharmonicity_cat.append(true_dataset[key][i]['analysis']['sfx']['inharmonicity']['mean'])

        except:
            #print("Id without duration: " + str(true_dataset[key][i]['id']))
            print("Id without some of the features: " + str(true_dataset[key][i]['id']))
    
    durations.append(durations_cat)
    silence_20.append(silence_20_cat)
    zcr.append(zcr_cat)
    logattacktime.append(logattacktime_cat)
    inharmonicity.append(inharmonicity_cat)
```

```python
durations = durations[1:]
#silence_20 = silence_20[1:]
zcr = zcr[1:]
logattacktime = logattacktime[1:]
inharmonicity = inharmonicity[1:]
```

## Durations

```python
# Create a figure instance
fig = plt.figure(1, figsize=(9, 6))

# Create an axes instance
ax = fig.add_subplot(111)

# Create the boxplot
bp = ax.boxplot(durations,patch_artist=True)
ax.set_xticklabels(['coi','cup','dor','dra','key','mic','pac','sci','toi','typ','vac','wrt','zip'])
ax.set_ylabel("s")
ax.set_xlabel("categories")


for box in bp['boxes']:
    # change outline color
    box.set( color='#7570b3', linewidth=2)
    # change fill color
    box.set( facecolor = '#1b9e77' )

for whisker in bp['whiskers']:
    whisker.set(color='#7570b3', linewidth=2)
    
for cap in bp['caps']:
    cap.set(color='#7570b3', linewidth=2)

for median in bp['medians']:
    median.set(color='#b2df8a', linewidth=2)

for flier in bp['fliers']:
    flier.set(marker='o', color='#e7298a', alpha=0.5)
```
